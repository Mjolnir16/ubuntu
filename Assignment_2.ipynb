```python
#No. of missing values in each column

df.isnull().sum()
```

```python
#dropping rows with missing values

df.dropna(inplace = False)
```

```python
#Filling missing values with mean

df.fillna(df.mean(), inplace = False)
```

```python
df
```

```python
#Filling missing values with mode

df_mode = df

for i in df_mode.columns:
    df_mode[i].fillna(df_mode[i].mode()[0], inplace = True)

df_mode
```

```python
#Filling missing values with median

df.fillna(df.median(), inplace = False)
```

```python
from scipy import stats
```

```python
z = np.abs(stats.zscore(df['HSC per']))
threshold = 3
outliers = df[z > threshold]
print(outliers)
```

```python
q1 = df['HSC per'].quantile(0.25)
q3 = df['HSC per'].quantile(0.75)

iqr = q3 - q1

higher = q3 + 1.5*iqr
lower = q1 - 1.5*iqr

filter1 = df['HSC per'] > higher
filter2 = df['HSC per'] < lower
outliers = df.where(filter1 | filter2, inplace = False)
print(outliers)
```

```python
df_removed = df
upper_array = np.where(df_removed['HSC per'] >= higher)[0]
lower_array = np.where(df_removed['HSC per'] <= lower)[0]
df_removed.drop(index = upper_array, inplace=True)
df_removed.drop(index = lower_array, inplace=True)

df_removed
```

```python
df = pd.read_csv('student_outliers.csv')
df["HSC per"] = np.where(df["HSC per"] > higher, df['HSC per'].mean(), df['HSC per'])
df
```

```python
df = pd.read_csv('student_outliers.csv')
df["HSC per"] = np.where(df["HSC per"] > higher, df['HSC per'].median(), df['HSC per'])
df
```
